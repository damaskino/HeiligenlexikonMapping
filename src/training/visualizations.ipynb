```python
import pandas as pd
```

```python
results_df = pd.read_csv('acc_results_across_sent_len_and_similarity_cutoff.csv', sep=";")
```

```python
df_list = []
for i in range(1,10):
    temp_df = results_df[results_df["SentenceLength"]==i]
    df_list.append(temp_df)
```

```python
for length_df in df_list:
    length_df.plot.bar(x="SimilarityThreshold", y="Accuracy")
```
